```python
# Exponential growth
exp_i = [1, 2, 4, 8, 16]
exp_r = [0, 1, 3, 7, 15, 31]
```

```python
plt.plot(t, i, 'b', label="Game")
plt.plot(t[:5], exp_i, 'r', label="Exponential growth")
plt.xlabel("Generation")
plt.ylabel("Tagged")
plt.xlim(0, len(t)-1)
plt.ylim(0, 16)
plt.legend()
```
