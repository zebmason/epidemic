```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Exponential growth
exp_i = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512]
exp_r = [0, 1, 3, 7, 15, 31, 63, 127, 255, 511, 1023]
```

```python
plt.plot(t, i, 'b', label="Game")
plt.plot(t[:10], exp_i, 'r', label="Exponential growth")
plt.xlabel("Generation")
plt.ylabel("Tagged")
plt.xlim(0, len(t)-1)
plt.ylim(0, 512)
plt.legend()
```

```python
plt.plot(t, r, 'b', label="Game")
plt.plot(t[:11], exp_r, 'r', label="Exponential growth")
plt.xlabel("Generation")
plt.ylabel("No longer tagged")
plt.xlim(0, len(t)-1)
plt.ylim(0, 1023)
plt.legend()
```
